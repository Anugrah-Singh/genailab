```python
import gensim.downloader as api
from transformers import pipeline


model = api.load("glove-wiki-gigaword-100")
print("Model loaded successfully")

generator = pipeline("text-generation", model = "gpt2")
```

```python
seedword = input("Enter the seedword")
similar_words = model.most_similar(seedword, topn=5)
print("Similar words:\n")
for w, score in similar_words:
    print(w, f"{score:.2f}")

extra_words = [w for w, _ in similar_words]

sentence=" ".join(extra_words)
print(sentence)

paragraph = f"{seedword.capitalize()} is an amazing worls. and these are related to it {",".join(extra_words[1:])}. these words togeter create a concept related to {seedword}"

print("\nGenerated Paragraph")
print(paragraph)
```
